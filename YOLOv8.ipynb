```python
import time
import os
from PIL import Image

# Path to the directory containing PNG images for inference
image_folder = '/home/pederu/Documents/YOLOv8_project/dataset/images/all'

# List all PNG files in the folder
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith('.PNG')]

# Prepare to collect inference times
total_inference_time = 0
num_images = len(image_files)

# Process each image and measure inference time
for image_path in image_files:
    # Load an image
    image = Image.open(image_path)

    # Start timing
    start_time = time.time()
    results = model.predict(image)
    end_time = time.time()

    # Calculate inference time for this image
    inference_time = end_time - start_time
    total_inference_time += inference_time

    # Optional: Print inference time for this image
    print(f"Inference time for {os.path.basename(image_path)}: {inference_time:.4f} seconds")

# Print average inference time
average_inference_time = total_inference_time / num_images
print(f"Average inference time: {average_inference_time:.4f} seconds")

# Print total inference time for all images
print(f"Total inference time for all images: {total_inference_time:.4f} seconds")
```
